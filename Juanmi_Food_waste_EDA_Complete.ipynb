```python

'''
 hay columnas que no nos sirven :
 "reference", "notes", "url", 
 "method_data_collection", 
 "loss_percentage_original")
'''
df2.columns
```

```python
# creamos una nueva df limpia y la llamamos df_clean:
df_clean = df2.drop(columns = ["reference", "notes", "url", "method_data_collection", "loss_percentage_original"])
df_clean
```

```python
#  Miramos las perdidas globales por año en las distintas etapas de la cadena de alimentos:
sns.set(style="whitegrid")

g = sns.catplot(data=df_clean, x='year', y='loss_percentage', hue='food_supply_stage', height=6, aspect=2)

g.set_axis_labels("Año", "Porcentaje de pérdida")
g.legend.set_title("Etapa de la cadena de alimentos")

sns.despine(left=True, bottom=True)
```

* Cultivo, cosecha, distribuciòn y transporte son las etapas que muestran mayor concentraciòn de desperdicio, este gráfico parece validar la hipótesis 4b. 

* Aunque la concentración general de desperdicio disminuye entre el 20-40%, el porcentaje general de desperdicio parece que sigue marcando por encima del 32% indicado por la FAO (Hipòtesis 1)





```python
# Ahora las causas de las pérdidas:
df_clean["cause_of_loss"].value_counts()
```

```python
# porcentaje de valores nulos de la causa de pèrdida de alimentos:
df_clean["cause_of_loss"].isnull().sum()/27773*100
```

96.67% de valores nulos, demasiaso alto, filtramos las 10 principales para ver si encontramos rechazo de calidad basado en apariencia estètica que justifique la hipotesis 3








```python
# Las 10 causas más comunes de la perdida de alimentos:

causas = df_clean["cause_of_loss"].value_counts(normalize=False)[:11]

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=causas.values, y=causas.index, hue=causas.index, palette=sns.color_palette("hls", len(causas)), legend=False)
plt.xlabel("Porcentaje %")
plt.ylabel("Causa de pérdida")
plt.title("Causas más comunes de pérdida de alimentos")
plt.show()





```

* Aunque tambien incluye pérdidas por desajustes entre oferta y demanda en el mismo apartado,las pérdidas debidas al aspecto estético aparecen en tercer lugar, una posiciòn bastante alta (40%)(H3)<br>

* Las Pérdidas físicas tampoco están muy claras de què se trata, puede que sean Pérdidas físicas
totales, incluyendo pérdidas físicas y mecánicas.


```python
# El porcentaje de media de pèrdida por paìs:
porcentaje_media = df_clean.groupby('country')['loss_percentage'].mean()

porcentaje_media

```

```python
# Miramos los 10 con màs pèrdida
top_10 = porcentaje_media.nlargest(10)

print(top_10)
```

```python
# Heat map con los 10 países con la mayor media pérdida en porcentaje de alimentos:
heatmap_data = df_clean[df_clean['country'].isin(top_10.index)].pivot_table(index='country', columns='year', values='loss_percentage')

sns.set(font_scale=1.2)
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', cbar_kws={'label': 'Loss Percentage'})
plt.title('Top 10 Countries with the Highest Loss Percentage')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

```python
# Parece que se ve mas claro con un barchart:
fig, ax = plt.subplots()
ax.barh(top_10.index, top_10.values, color='blue')
ax.set_xlabel('Porcentaje %')
ax.set_title('Países con la mayor pérdida de alimentos')

plt.show()
```

```python
# los cinco paises que màs desperdician y en què estado de la cadena?:

top_5 = df_clean[df_clean["country"].isin(["Haiti", "Australia and New Zealand", "Oman", "Gabon", "Africa"])]
top_group = top_5.groupby(["food_supply_stage", "country"])["loss_percentage"].mean().reset_index()

sns.barplot(data=top_group, y='country', x='loss_percentage', hue='food_supply_stage')
plt.xlabel("Porcentaje de perdida")
plt.title("Los cinco paises que mas gastan y en què estado de la cadena?")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()

```

```python
# Los 10 países con la menor media de pérdida en porcentaje de alimentos:
bottom_10 = df_clean.groupby('country')['loss_percentage'].mean().nsmallest(10)
bottom_10
```

```python
# Horizontal bar plot de los 10 países con la menor media de pérdida en porcentaje de alimentos:
fig, ax = plt.subplots()
ax.barh(bottom_10.index, bottom_10.values, color='green')
ax.set_xlabel('Porcentaje %')
ax.set_title('Países con la menor pérdida de alimentos')

plt.show()





```

```python
#  Los diez países que menos desperdician alimentos y estado de la cadena 
bottom_10 = df_clean.groupby('country')['loss_percentage'].mean().sort_values().head(10)
bottom_10_df = df_clean[df_clean['country'].isin(bottom_10.index)]
sns.barplot(x='loss_percentage', y='country', hue='food_supply_stage', data=bottom_10_df)
plt.xlabel("Porcentaje de pérdida")
plt.title("Los diez países que menos desperdician alimentos")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
